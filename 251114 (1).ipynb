## Pandas 주요 함수 및 메소드 정리

Pandas는 데이터 분석에 필수적인 Python 라이브러리입니다. 다음은 가장 자주 사용되는 DataFrame 및 Series의 함수와 메소드들을 카테고리별로 정리한 내용입니다.

### 1. 데이터 불러오기 및 저장

*   `pd.read_csv()`: CSV 파일을 DataFrame으로 불러옵니다.
*   `pd.read_excel()`: Excel 파일을 DataFrame으로 불러옵니다.
*   `pd.read_sql()`: SQL 쿼리 결과를 DataFrame으로 불러옵니다.
*   `df.to_csv()`: DataFrame을 CSV 파일로 저장합니다.
*   `df.to_excel()`: DataFrame을 Excel 파일로 저장합니다.

### 2. 데이터 탐색 및 요약

*   `df.head(n)`: DataFrame의 상위 `n`개 행을 반환합니다. (기본값 5)
*   `df.tail(n)`: DataFrame의 하위 `n`개 행을 반환합니다. (기본값 5)
*   `df.info()`: DataFrame의 요약 정보를 출력합니다. (컬럼별 non-null 개수, Dtype 등)
*   `df.describe()`: DataFrame의 숫자형 컬럼에 대한 기술 통계량(평균, 표준편차, 최소값, 최대값 등)을 계산합니다.
*   `df.shape`: DataFrame의 행과 열의 개수를 튜플로 반환합니다.
*   `df.columns`: DataFrame의 컬럼 이름을 반환합니다.
*   `df.index`: DataFrame의 인덱스를 반환합니다.
*   `df.value_counts()`: Series의 고유한 값들과 그 개수를 반환합니다.
*   `df.unique()`: Series의 고유한 값들을 배열로 반환합니다.

### 3. 데이터 선택 및 필터링

*   `df['컬럼명']`: 단일 컬럼을 Series로 선택합니다.
*   `df[['컬럼1', '컬럼2']]`: 여러 컬럼을 DataFrame으로 선택합니다.
*   `df.loc[행_라벨, 열_라벨]`: 라벨(이름)을 사용하여 데이터에 접근합니다.
    *   예: `df.loc[0, '컬럼명']`, `df.loc[0:5, ['컬럼1', '컬럼2']]`
*   `df.iloc[행_위치, 열_위치]`: 정수 위치를 사용하여 데이터에 접근합니다.
    *   예: `df.iloc[0, 0]`, `df.iloc[0:5, 0:2]`
*   `df[조건]`: 조건에 맞는 행을 필터링합니다.
    *   예: `df[df['컬럼명'] > 10]`
*   `df.query('컬럼명 > 10')`: 문자열 쿼리를 사용하여 데이터를 필터링합니다.

### 4. 데이터 정렬

*   `df.sort_values(by='컬럼명', ascending=True)`: 특정 컬럼을 기준으로 DataFrame을 정렬합니다. `ascending=False`는 내림차순 정렬입니다.
*   `df.sort_index(ascending=True)`: 인덱스를 기준으로 DataFrame을 정렬합니다.

### 5. 결측치(Missing Values) 처리

*   `df.isnull()`: DataFrame의 각 요소가 결측치(NaN, None)인지 여부를 나타내는 불리언 DataFrame을 반환합니다.
*   `df.isna()`: `isnull()`과 동일합니다.
*   `df.notnull()`: DataFrame의 각 요소가 결측치가 아닌지 여부를 나타내는 불리언 DataFrame을 반환합니다.
*   `df.dropna(axis=0, how='any')`: 결측치가 있는 행(`axis=0`) 또는 열(`axis=1`)을 제거합니다. `how='all'`은 모든 값이 결측치인 경우에만 제거합니다.
*   `df.fillna(value)`: 결측치를 특정 값으로 채웁니다.
    *   예: `df.fillna(0)`, `df.fillna(df.mean())`

### 6. 데이터 타입(Dtype) 변환

*   `df['컬럼명'].astype(new_type)`: 컬럼의 데이터 타입을 변경합니다.
    *   예: `df['나이'].astype(int)`, `df['날짜'].astype('datetime64[ns]')`
*   `pd.to_datetime(series)`: Series를 datetime 객체로 변환합니다.
*   `pd.to_numeric(series, errors='coerce')`: Series를 숫자형으로 변환합니다. 변환할 수 없는 값은 `NaN`으로 처리합니다.

### 7. 컬럼 추가, 삭제, 이름 변경

*   `df['새_컬럼'] = 값`: 새로운 컬럼을 추가합니다.
*   `df.drop('컬럼명', axis=1, inplace=True)`: 특정 컬럼을 제거합니다. `inplace=True`는 원본 DataFrame을 변경합니다.
*   `df.rename(columns={'옛_이름': '새_이름'})`: 컬럼 이름을 변경합니다.

### 8. 데이터 그룹화 및 집계

*   `df.groupby('컬럼명')`: 특정 컬럼을 기준으로 데이터를 그룹화합니다.
    *   그룹화 후 `sum()`, `mean()`, `count()`, `min()`, `max()`, `agg()` 등의 집계 함수를 사용할 수 있습니다.
    *   예: `df.groupby('도시')['판매량'].sum()`
*   `df.pivot_table()`: 데이터를 피벗 테이블 형태로 재구성하고 집계합니다.

### 9. 데이터 병합 (Merge/Join)

*   `pd.merge(df1, df2, on='키_컬럼', how='inner')`: 두 DataFrame을 특정 키 컬럼을 기준으로 병합합니다. `how` 인자에는 `inner`, `outer`, `left`, `right`가 있습니다.
*   `pd.concat([df1, df2])`: 두 DataFrame을 행 또는 열 방향으로 이어 붙입니다.

### 10. 문자열 처리 (Series.str)

*   `df['컬럼'].str.contains('문자열')`: 특정 문자열을 포함하는지 확인합니다.
*   `df['컬럼'].str.replace('old', 'new')`: 문자열을 바꿉니다.
*   `df['컬럼'].str.split('구분자')`: 문자열을 분리합니다.
*   `df['컬럼'].str.upper()`, `df['컬럼'].str.lower()`: 대문자/소문자로 변환합니다.
*   `df['컬럼'].str.strip()`: 공백을 제거합니다.

### 11. 날짜/시간 처리 (Series.dt)

*   `df['날짜컬럼'].dt.year`: 연도를 추출합니다.
*   `df['날짜컬럼'].dt.month`: 월을 추출합니다.
*   `df['날짜컬럼'].dt.day`: 일을 추출합니다.
*   `df['날짜컬럼'].dt.dayofweek`: 요일(0=월요일, 6=일요일)을 추출합니다.
*   `df['날짜컬럼'].dt.weekday`: `dayofweek`와 동일합니다.
