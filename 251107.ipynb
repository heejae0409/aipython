```python
# 드론이 비행할 때, 이동 거리와 풍속에 따라 배터리 소모량이 어떻게 달라지는지 계산
distance = np.array([1, 2, 3, 5, 8])
windspeed = np.array([0, 5, 10, 7, 12])
basic = 10
additional = windspeed
total = basic + additional
battery = distance * total
print("비행 거리 (km):", distance)
print("풍속 (m/s):", windspeed)
print("총 소모율 (%/km):", total)
print("예상 배터리 소모량 (%):", battery)
```
